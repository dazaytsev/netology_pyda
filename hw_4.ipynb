```python
def _fibosum(n):
    f_num = [0 ,1]
    v_tmp, v_1, v_2 = 0, 0, 1

    if n < 3:
        print('Числа Фибоначчи: ', f_num)
        return sum(f_num[:n])
    if n > 2:
        for a in range(3, n+1):
            v_tmp = v_2     #промежуточная переменная
            v_2 = v_1 + v_2 #значение следующего числа в ряду
            v_1 = v_tmp     #значение предыдущего числа в ряду
            f_num.append(v_2)
        print('Числа Фибоначчи: ', f_num)
        return sum(f_num[:n])
```

К условию из задания 5 реализуйте следующие команды:  
  
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;

- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
documents = [{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
             {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
             {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]
directories = {'1': ['2207 876234', '11-2', '5455 028765'],
               '2': ['10006', '5400 028765', '5455 002299'],
               '3': []}

#запрос номера документа
def _get_doc_num():
    #ret_vl = input('Введите номер документа: ')
    ret_vl = '2207 876234'
    return ret_vl

#запрос номера полки
def _get_sh_num():
    #ret_vl = input('Введите номер полки: ')
    ret_vl = '3'
    return ret_vl

#запрос данных
def _get_vl(q_str):
    #ret_vl = input(q_str)
    #return ret_vl
    return input(q_str)

#проверка наличия документа
def _check_doc(doc_num):
    ret_vl = False
    for d in documents:
        if doc_num == d['number']:
            ret_vl = True
    return ret_vl

#проверка наличия полки
def _check_st(s_num):
    ret_vl = False
    for d in directories.keys():
        if s_num == d:
            ret_vl = True
    return ret_vl

#p - поиск человека
def _get_people():
    doc_num = _get_vl('Укажите номер документа: ')
    
    ret_vl = 'Введен несуществующий документ'
    for d in documents:
        if doc_num == d['number']:
            ret_vl = d['name']
    print(ret_vl)

#s - поиск полки
def _get_directory():
    doc_num = _get_vl('Укажите номер документа: ')
    
    ret_vl = 'Введен несуществующий документ'
    for key, value in directories.items():
        if doc_num in value:
            ret_vl = key
    print(ret_vl)

#l - список документов
def _view_all_docs():
    for d in documents:
        print(d['type'], '\"' + d['number'] + '\"', '\"' + d['name'] + '\"')

#as - новый номер полки
def _add_new_sh():
    st_num = _get_vl('Укажите номер полки для добавления: ')
    
    if st_num not in directories.keys(): 
        directories[st_num] = []
    else:
        print("Ошибка добавления: полка с номером " + st_num + " уже существует")

#d - удаление записи
def _del_doc():
    doc_num = _get_vl('Укажите номер документа для удаления: ')
    
    if _check_doc(doc_num):
        for d in documents:
            if doc_num == d['number']:
                documents.remove(d)
                ret_vl = 'Запись для документа удалена'
                break

        for d in directories:
            if doc_num in directories[d]:
                directories[d].remove(doc_num)
                break
    else:
        ret_vl = 'Ошибка удаления: попытка удалить несуществующий документ'
    print(ret_vl)

#m - перенос
def _migr_doc():
    st_num = _get_vl('Укажите номер полки для хранения документа: ')
    doc_num = _get_vl('Укажите номер документа: ')
    
    msg_doc = _check_doc(doc_num)
    msg_s = _check_st(st_num)
    
    if not (msg_doc and msg_s):
        print ('ошибка')
    else:
        for d in directories:
            if doc_num in directories[d]:
                directories[d].remove(doc_num)
                directories[st_num].append(doc_num)

#a - добавление новой записи
def _add_new():
    st_num = _get_vl('Укажите номер полки для хранения документа: ')
    if not _check_st(st_num):
        print('Указана несуществующая полка')
        return
    
    doc_num = _get_vl('Укажите номер документа: ')
    if _check_doc(doc_num):
        print('Указан существующий номер документа')
        return
    
    doc_type = _get_vl('Укажите тип документа: ')
    nm = _get_vl('Укажите имя владельца: ')
    
    msg_doc = _check_doc(doc_num)
    msg_s = _check_st(st_num)
    
    documents.append({'type': doc_type, 'number': doc_num, 'name': nm})
    directories[st_num].append(doc_num)
```
